```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('round-1-island-data-bottle/prices_round_1_day_-2.csv', delimiter=';')

# Display the first few rows of the DataFrame
df.head()
```

```python
df_ink = df[df['product'] == 'SQUID_INK']
df_ink.head()
df_ink.tail()
df_ink.describe()
df_ink.info()
df_ink.isnull().sum()
df_ink.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Plot for RAINFOREST_RESIN
plt.figure(figsize=(12, 6))
plt.plot(df_ink['timestamp'], df_ink['mid_price'], label='SQUID_INK')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.title('Mid Price for SQUID_INK')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import pandas as pd

# Load the CSV file
df_fair = pd.read_csv('round-1-island-data-bottle/IMC Prosperity Results (15).csv', delimiter=';')
df_fair = df_fair[df_fair['product'] == 'SQUID_INK']
# Display the first few rows of the DataFrame
df_fair.head()
```

```python
import matplotlib.pyplot as plt

# Plot for RAINFOREST_RESIN
plt.figure(figsize=(12, 6))
plt.plot(df_fair['timestamp'], df_fair['mid_price'], label='SQUID_INK')
plt.plot(df_fair['timestamp'], 1840+df_fair['profit_and_loss'], label='SQUID_INK')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.title('Mid Price for SQUID_INK')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the adjusted P&L and the difference
adjusted_pnl = 1840 + df_fair['profit_and_loss']
difference = df_fair['mid_price'] - adjusted_pnl

plt.figure(figsize=(12, 6))
plt.plot(df_fair['timestamp'], df_fair['mid_price'], label='Mid Price')
plt.plot(df_fair['timestamp'], difference, label='Difference', linestyle='--')

plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Mid Price vs Adjusted P&L and Their Difference')
plt.legend()
plt.grid(True)
plt.show()
```
